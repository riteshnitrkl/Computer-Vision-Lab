```python
import cv2
import numpy as np

# Load the images
image1 = cv2.imread('lion.jpg')
image2 = cv2.imread('Elephant.jpg')


if image1 is None or image2 is None:
    print("Error: One or both images not found.")
    exit()


height1, width1 = image1.shape[:2]
height2, width2 = image2.shape[:2]


new_height = 400  

image1_resized = cv2.resize(image1, (int(width1 * new_height / height1), new_height))
image2_resized = cv2.resize(image2, (int(width2 * new_height / height2), new_height))


combined_image = np.hstack((image1_resized, image2_resized))

cv2.imshow('Combined Image Lion Elephant', combined_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

image = cv2.imread('Elephant.jpg')

if image is None:
    print("Error: Image not found.")
    exit()

cv2.imshow('Original Image', image)

start_x = 100  
start_y = 50   
region_width = 150
region_height = 150

end_x = min(start_x + region_width, image.shape[1])
end_y = min(start_y + region_height, image.shape[0])

extracted_region = image[start_y:end_y, start_x:end_x]

cv2.imshow('Extracted Region', extracted_region)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('Elephant.jpg')

if image is None:
    print("Error: Image not found.")
    exit()

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


hist = cv2.calcHist([gray_image], [0], None, [256], [0, 256])

plt.figure(figsize=(8, 6))
plt.plot(hist, color='black')
plt.title('Grayscale Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.grid(True)

output_file = 'grayscale_histogram.png'
plt.savefig(output_file)

print(f"Histogram saved as '{output_file}'.")

plt.show()
```

```python
import cv2
import numpy as np

image = cv2.imread('Elephant.jpg')


if image is None:
    print("Error: Image not found.")
    exit()

gaussian_filter = np.array([
    [1, 2, 1],
    [2, 4, 2],
    [1, 2, 1]
], dtype=np.float32) / 16 


blurred_image = cv2.filter2D(image, -1, gaussian_filter)


cv2.imshow('Original Image', image)
cv2.imshow('Blurred Image', blurred_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
